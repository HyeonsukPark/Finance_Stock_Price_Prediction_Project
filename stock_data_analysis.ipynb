## Moving Averages

Moving Average is a stock indicator that helps traders and inverstors identify the direction of a price trend
by smoothing out the day to day noise and fluctuations in the stock's price data. 

Moving Average is a continuously updated average price, calculated over a specific number of time periods, 
with the oldest data point dropping off as the newest one is added. 

* if the MA line is rising, it confirms an uptrend
* if the MA line is falling, it confirms a downtrend 

1. Simple Moving Average (SMA) 
This is calculated by taking the arithmetric mean of a set of prices over a specified period. 
SMA is used to identify stable, long-term trends, but it is slow to react to sharp, recent price changes. 

2. Exponential Moving Average (EMA) 
It uses a complex formula that applies a higher weighting to the most recent price data and less weight 
to older data. EMA is more sensitive and reacts to current price changes than SMA, making it popular for short
-term traders.




```python
# 20-Day SMA (for short/medium term trend) 
# 'rolling(window=20).mean()' function calculates the average of the previous 20 periods.

# 50-Day EMA (often used for medium/long-term trend)
# The 'ewm(span=50, adjust=False).mean()' function gives more weight to recent prices.

# Tesla
tsla['SMA_20'] = tsla['Close'].rolling(window=20).mean()
tsla['EMA_50'] = tsla['Close'].ewm(span=50, adjust=False).mean()

nvda['SMA_20'] = nvda['Close'].rolling(window=20).mean()
nvda['EMA_50'] = nvda['Close'].ewm(span=50, adjust=False).mean()

amzn['SMA_20'] = amzn['Close'].rolling(window=20).mean()
amzn['EMA_50'] = amzn['Close'].ewm(span=50, adjust=False).mean()

aapl['SMA_20'] = aapl['Close'].rolling(window=20).mean()
aapl['EMA_50'] = aapl['Close'].ewm(span=50, adjust=False).mean()

nflx['SMA_20'] = nflx['Close'].rolling(window=20).mean()
nflx['EMA_50'] = nflx['Close'].ewm(span=50, adjust=False).mean()
```

Stock Close Price : The actual closing price of stock on each trading day

20-day SMA 
* the average closing price over the last 20 days. It reacts more quickly to price changes. 
* Identify short-term trends.

50-day EMA 
* A weighted average over the last 50 days, giving more importance to recent prices. 
* It reacts more slowly than the 20 day SMA. 
* Identify medium-term trends.

## Prediction Pipeline 

```python
# Stock dataframe 

stock_data = {
    'TSLA': tsla, 
    'NVDA': nvda,
    'AMZN': amzn,
    'AAPL': aapl,
    'NFLX': nflx
}
```

```python
# LSTM dataset 
# converting a list of prices into sequences 
# look_back is 60. the model will learn to predict P(61) from the sequence P(1) to P(60)

def create_lstm_dataset(data, look_back=60): 
    X, Y = [], []
    for i in range(len(data) - look_back - 1):
        a = data[i: (i + look_back), 0]
        X.append(a)
        Y.append(data[i + look_back, 0])
    return np.array(X), np.array(Y)
```

```python
# Training function

def build_and_train_lstm(df_stock, ticker, look_back=60):
    
    print(f"--- Starting Pipeline for {ticker} ---")
    # prepare data and scale 
    data = df_stock['Close'].values.reshape(-1, 1)
    
    # Scale
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)
    
    # split into training and testing sets (80% train, 20% test)
    training_data_len = int(len(scaled_data) * 0.8)
    train_data = scaled_data[0:training_data_len, :]
    
    # Create the training dataset (X = 60 previous days, Y = 61st day)
    X_train, y_train = create_lstm_dataset(train_data, look_back)
    
    # reshape the input data for LSTM (samples, time steps, features)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    
    # ----------------------------------------------------------------------
    
    # build LSTM model
    model = Sequential()
    model.add(LSTM(units=100, return_sequences=True, input_shape=(1, 1)))
    model.add(LSTM(units=100, return_sequences=False))
    model.add(Dense(25))
    model.add(Dense(1))
    
    # Compile and Train
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0)
    
    # Create the test set
    test_data = scaled_data[training_data_len - look_back:, :]
    X_test, y_test = create_lstm_dataset(test_data, look_back)
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
    
    # Predictions 
    predictions = model.predict(X_test, verbose=0)
    
    predictions_unscaled = scaler.inverse_transform(predictions)
    y_test_unscaled = scaler.inverse_transform(y_test.reshape(-1, 1))
    
    # ----------------------------------------------------------------------
    
    # metrics 
    # MSE 
    mse = mean_squared_error(y_test_unscaled, predictions_unscaled)
    
    # RMSE
    rmse = np.sqrt(mse)
    
    print(f" Metrics for {ticker}:")
    print(f" Mean Squared Error (MSE): {mse:.4f}")
    print(f" Root Mean Squared Error (RMSE): {rmse:.4f}")
    
    # ----------------------------------------------------------------
    
    # Plotting results
    train = df_stock[:training_data_len].copy()
    
    valid_start_index = len(df_stock) - len(predictions_unscaled)
    valid = df_stock[valid_start_index:].copy()
    
    valid['Predictions'] = predictions_unscaled
    
    plt.figure(figsize=(12, 6))
    plt.title(f'LSTM Prediction for {ticker} (RMSE: {rmse:.2f})')
    plt.xlabel('Date')
    plt.ylabel('Close Price ($)')
    
    # plot training data
    plt.plot(train.index, train['Close'], label='Training Data', color='blue')
    
    # Plot actual test data and predictions
    plt.plot(valid.index, valid['Close'], label='Actual Value', color='red')
    plt.plot(valid.index, valid['Predictions'], label='Predictions', color='green')
    
    plt.legend(loc='lower right')
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.tight_layout()
    
    return valid
    
    
```

### Running the function

```python
results = {}

for ticker, df in stock_data.items():
    result_df = build_and_train_lstm(df, ticker)
    results[ticker] = result_df
```
